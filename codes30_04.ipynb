```python
#5
s = input()
if "+" in s:
    print(int(s[:s.find("+")]) + int(s[s.find("+")+1:]))
elif "-" in s:
    print(int(s[:s.find("-")]) - int(s[s.find("-")+1:]))
elif "/" in s:
    print(int(s[:s.find("/")]) / int(s[s.find("/")+1:]))
elif "*" in s:
    print(int(s[:s.find("*")]) * int(s[s.find("*")+1:]))
```

```python
#7
a,b = int(input("от")), int(input("до "))
f = 0
for i in range(a,b+1):
    for y in range(1,i):
        if i % y == 0:
            f += 1
    if f == 1:
        print(i)
    f = 0
```

```python
#9
a = input()
a1 = a
max,prevmax = 0,0
for i in range(a.count("0")):
    n = a.find("0")
    if int(a[0:n]) > max:
        max = int(a[0:n])
    a = a[n+1:]
a2 = a1.replace(str(max)+"0","")
for i in range(a2.count("0")):
    n = a2.find("0")
    if int(a2[0:n]) > prevmax:
        prevmax = int(a2[0:n])
    a2 = a2[n+1:]
print(prevmax)
```

```python
#3
a1, a2 = input(), input()
alph = set('abcdefghijklmnopqrstuvwxyz')
#Уникальные символы:
a1, a2 = set(a1), set(a2)
#Общие:
a = a1 & a2
print(*a)
#Есть только в первой:
a11 = a1 - a2
print(*a11)
#Есть только во второй:
a22 = a2 - a1
print(*a22)
#Есть в алфавите, нет в строках:
print(*(alph - (a1 | a2)))
```
